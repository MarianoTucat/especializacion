# Qué es DynamoDB

- Base de datos NoSQL proprietaria de AWS, que ofrece:
  - performance confiable, que se mantiene si escala su uso
  - serverless (no nos tenemos que preocupar por donde corre)
  - acceso via API simple y pequeña, para casos simples de clave valor, pero también para consultas avanzadas


# En que casos es conveniente

- Aplicaciones con grandes tamaños de información y requerimientos estrictos de latencia
- Aplicaciones serverless usando AWS Lambda
- Para data sets simples, con patrones de accesos conocidos

# Claves Primarias

# Indices Secundarios

- local secondary index (comparten la clave de partición de la clave primaria, cambian la clave de ordenamiento)
- global secondary index (definen una nueva clave de partición y potencialmente una nueva clave de ordenamiento)

# Capacidad de Lectura y Escritura (RCU & WCU)

En dynamo es clave especificar correctaente las unidades de capacidad para lectura y escritura.
Esto nos permite un cierto número de operaciones por segundo, además del costo que tendrá su uso.
Además, dynamo provee autoscaling de estas unidades, para permitir ajustarse a horarios picos del uso.
